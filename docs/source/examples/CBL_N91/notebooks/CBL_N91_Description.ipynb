**Case Description**

This intercomparison case (henceforth CBL_N91) simulates a (free) convective boundary layer driven by a constant surface sensible heat flux. The flow develops a deepening, well-mixed convective layer with no geostrophic wind.

| Parameter | Value                                                                           |
| --- |---------------------------------------------------------------------------------|

| Domain | 6400 m x 6400 m x 5000 m                                                        |
| Reference grid | $128^3$                                                                         |
| Run matrix | $64^3$, $128^3$, $256^3$, $384^3$ x LASDD-SM, LASDD-WL, LAD-SM, LAD-WL x SP, DP |
| Simulation time | 12600 s (3.5 h)                                                                 |
| Reference time step | 0.25 s                                                                          |
