```python
def LoadStats(StatFiles):
    
    # Initialize lists
    U   = []
    V   = []
    TH  = []
    u2  = []
    v2  = []
    w2  = []
    TH2 = []
    uv  = []
    uw  = []
    vw  = []
    txy = []
    txz = []
    tyz = []
    
    # Process each file
    for file in StatFiles:
        with np.load(file) as data:
            U.append(data['u'])
            V.append(data['v'])
            TH.append(data['TH'])
            u2.append(data['u2'])
            v2.append(data['v2'])
            w2.append(data['w2'])
            TH2.append(data['TH2'])
            uv.append(data['uv'])
            uw.append(data['uw'])
            vw.append(data['vw'])
            txy.append(data['txy'])
            txz.append(data['txz'])
            tyz.append(data['tyz'])
    
    # Convert lists to numpy arrays
    U = np.array(U)
    V = np.array(V)
    TH = np.array(TH)
    u2 = np.array(u2)
    v2 = np.array(v2)
    w2 = np.array(w2)
    TH2 = np.array(TH2)
    uv = np.array(uv)
    uw = np.array(uw)
    vw = np.array(vw)
    txy = np.array(txy)
    txz = np.array(txz)
    tyz = np.array(tyz)
    
    return (U, V, TH, u2, v2, w2, TH2, uv, uw, vw, txy, txz, tyz)
```
