**Case Description**

The GABLS1 benchmark is a moderately stable boundary layer driven by a prescribed surface
cooling rate of 0.25 K hr$^{-1}$ applied for 9 hours. The flow develops from an initially
neutral 265 K boundary layer (capped by an inversion above 100 m) and reaches a quasi-steady
stable state during the final hour.

| Parameter | Value |
| --- | --- |
| Domain | 400 m x 400 m x 400 m |
| Grid (reference run) | $40^3$ |
| Simulation time | 9 h |
| Geostrophic wind | $U_g = 8$ m s$^{-1}$, $V_g = 0$ m s$^{-1}$ |
| Coriolis parameter | $f = 1.39 \times 10^{-4}$ s$^{-1}$ (lat 73$^\circ$ N) |
| Surface cooling rate | 0.25 K hr$^{-1}$ |
| Initial surface temperature | $T_0 = 265$ K |
| Inversion strength | 0.01 K m$^{-1}$ above 100 m |
| Roughness length | $z_{0m} = z_{0T} = 0.1$ m |
| SGS model | LASDD-SM (`optSgs = 1`) |
| Damping layer | above $z = 300$ m |

The reference run shown below is `examples/SBL_GABLS1/runs/40x40x40_LASDD_SM_SP`.
